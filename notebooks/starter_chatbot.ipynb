```python
from src.schema import ProblemState, Supplier, Consumer, Node, Product
```

```python
state = ProblemState(
    problem_title="Simple Supply Chain Example"
)

state.nodes.append(Node(name="CountyA"))
state.nodes.append(Node(name="CountyB"))

state.products.append(Product(name="manure"))

state.suppliers.append(
    Supplier(
        name="Farm1",
        node="CountyA",
        product="manure",
        capacity=100,
        bid=0
    )
)

state.consumers.append(
    Consumer(
        name="Field1",
        node="CountyB",
        product="manure",
        capacity=80,
        bid=10
    )
)

state
```

```python
print("Nodes:", state.nodes)
print("Suppliers:", state.suppliers)
print("Consumers:", state.consumers)
```
